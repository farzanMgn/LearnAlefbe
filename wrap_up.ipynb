```python
df['tutor_payout,_pct'] = (
    df['tutor_payout,_pct']
    .astype(str)
    .str.replace('%', '', regex=False)
    .str.strip()
    .pipe(pd.to_numeric, errors='coerce')
)
```

```python
df['earning,_usd'] = (
    df['earning,_usd']
    .astype(str)
    .str.replace('[$,]', '', regex=True)
    .str.strip()
    .pipe(pd.to_numeric, errors='coerce')
)
```
