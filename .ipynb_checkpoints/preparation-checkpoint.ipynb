```python
df['tutor_payout,_pct'] = (
    df['tutor_payout,_pct']
    .astype(str)
    .str.replace('%', '', regex=False)
    .str.strip()
    .pipe(pd.to_numeric, errors='coerce')
)
```

```python
df['earning,_usd'] = (
    df['earning,_usd']
    .astype(str)
    .str.replace('[$,]', '', regex=True)
    .str.strip()
    .pipe(pd.to_numeric, errors='coerce')
)
```

```python
print("── Column Data Types ────────────────────────────────────────────────────")
print(df.dtypes)

print("\n── Shape ────────────────────────────────────────────────────────────────")
print(f"Rows: {df.shape[0]:,}  |  Columns: {df.shape[1]}")
```

```python
df['lesson_date'].fillna(pd.Timestamp('2026-05-17'), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# See all rows where student name is "."
dot_students = df[df['student'] == ' .']
print(dot_students[['student', 'student_location', 'lesson_price,_usd']].drop_duplicates())
print("\nTotal rows:", len(dot_students))
```

```python
df.head(10)
```

```python
mask = df['student'] == ' .'

df.loc[mask, 'student'] = df.loc[mask].groupby(
    ['student_location', 'lesson_price,_usd']
).ngroup().add(1).apply(lambda x: f'Unknown_{x}')

# Verify
print(df[df['student'].str.startswith('Unknown')][['student', 'student_location', 'lesson_price,_usd']].drop_duplicates().sort_values('student'))
```

```python
print(df['student_location'].unique())
print("\nTotal unique locations:", df['student_location'].nunique())
```

```python
print(df[df['student_location'] == 'PR'][['student', 'student_location', 'lesson_price,_usd', 'academic_year']])
```

```python
df['student_location'] = df['student_location'].str.replace('PR', 'Puerto Rico', regex=False)
```

```python
print(df['student_location'].value_counts())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
for col in ['student', 'student_location', 'type', 'academic_year']:
    df[col] = df[col].str.strip().str.title()

print(df['type'].unique())
print(df['academic_year'].unique())
```

```python
# Collapse all Unused into one category
df['type'] = df['type'].str.replace(r'Unused Lessons?\s*\(.*\)', 'Unused Lessons', regex=True)
```

```python
print(df['type'].unique())
print(df['academic_year'].unique())
```

```python
# Drop only the one real duplicate
df.drop_duplicates(inplace=True)
```
