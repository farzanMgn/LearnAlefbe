```python
import pandas as pd
import glob
import os
```

```python
folder_path = "./dataset" 
```

```python
files = sorted(glob.glob(os.path.join(folder_path, "tutor_report_*.csv")))
print(f"Found {len(files)} files:")
for f in files:
    print(f"  ✔ {os.path.basename(f)}")
```

```python
dfs = []
for f in files:
    year_label = os.path.basename(f).replace("tutor_report_", "").replace(".csv", "")
    df = pd.read_csv(f)
    df["academic_year"] = year_label          # adds a year column so you always know the source
    dfs.append(df)
```

```python
all_data = pd.concat(dfs, ignore_index=True)

# ── 4. Standardize column names (lowercase, underscores, no spaces) ───────────
all_data.columns = (
    all_data.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("%", "pct")
)
```

```python
print(f"\n✅ Combined dataset: {all_data.shape[0]:,} rows × {all_data.shape[1]} columns")
print(f"   Academic years:   {sorted(all_data['academic_year'].unique())}")
print("\n── Column names ──────────────────────────────────────────────────────────")
print(all_data.columns.tolist())
print("\n── First 5 rows ──────────────────────────────────────────────────────────")
all_data.head()
```

```python
output_path = os.path.join(folder_path, "tutor_report_ALL_YEARS.csv")
all_data.to_csv(output_path, index=False)
print(f"💾 Saved → {output_path}")
```
